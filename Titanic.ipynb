Data Preprocessing

```python
titanic.isna().sum()
```

```python
sns.heatmap(titanic.isna())
```

```python
titanic.drop('Cabin',axis=1,inplace=True)
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
```

```python
titanic.isna().sum()
```

```python
sns.heatmap(titanic.isna())
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic['Sex'].value_counts()
```

```python
titanic['Embarked'].value_counts()
```

```python
sex=pd.get_dummies(titanic['Sex'])
sex
```

```python
emb=pd.get_dummies(titanic['Embarked'])
emb
```

```python
titanic=pd.concat([titanic,sex,emb],axis=1)
titanic.head()
```

```python
titanic['female'] = titanic['female'].replace({'False':0,'True':1}).astype(np.uint8)
titanic['male'] = titanic['male'].replace({'False':0,'True':1}).astype(np.uint8)
titanic['C'] = titanic['C'].replace({'False':0,'True':1}).astype(np.uint8)
titanic['Q'] = titanic['Q'].replace({'False':0,'True':1}).astype(np.uint8)
titanic['S'] = titanic['S'].replace({'False':0,'True':1}).astype(np.uint8)
titanic
```

```python
titanic.drop('PassengerId',axis=1,inplace=True)
```

```python
titanic.drop('Sex',axis=1,inplace=True)
```

```python
titanic.drop('Name',axis=1,inplace=True)
```

```python
titanic.drop('Ticket',axis=1,inplace=True)
```

```python
titanic.drop('Embarked',axis=1,inplace=True)
```

```python
titanic.drop('female',axis=1,inplace=True)
```

Model Traning

```python
x=titanic.drop('Survived',axis=1)
y=titanic['Survived']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=5000)
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.svm import SVC
```

```python
model1=SVC(max_iter=5000)
model1.fit(x_train,y_train)
```

```python
y_predict1=model1.predict(x_test)
```

```python
print(classification_report(y_test,y_predict1))
```
